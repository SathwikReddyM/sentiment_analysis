```python
import pandas as pd
data = pd.read_csv("Data/train_data.csv")
data = data.drop("Unnamed: 0", axis=1)
data
```

```python

import spacy
nlp = spacy.load('en_core_web_sm')
processed_data = []
for i in range(len(data)):
    x = []
    for token in nlp(data.iloc[i]["Caption"]):
        if not token.like_email and not token.is_punct and not token.like_url and not token.is_stop and not token.is_space:
            x.append(token.lemma_)
    processed_data.append(x)
processed_data
            
            
```

```python
data["processed_tweets"] = processed_data
data
```
